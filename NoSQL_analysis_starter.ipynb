### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"There are {count} establishments that has a scored a 20 in the Hygiene category.")

# Display the first document in the results using pprint
pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(establishments.find(query))

# Display the number of rows in the DataFrame
print(f"There are {len(df)} rows within this dataframe.")

# Display the first 10 rows of the DataFrame
df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
establishments.find_one({"AddressLine3": "London"})
```

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'LocalAuthorityName': {'$regex': 'London'},
         'RatingValue': {'$gte': '4'}}
           
# Use count_documents to display the number of documents in the result
londoncount = establishments.count_documents(query)
print(f"There are {londoncount} restraurants in London that scored 4 and above in ratings.")

# Display the first document in the results using pprint
pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
londonestablishment = pd.DataFrame(establishments.find(query))

# Display the number of rows in the DataFrame
print(f"{len(londonestablishment)} rows are within the London dataframe.")

# Display the first 10 rows of the DataFrame
df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
pprint(establishments.find_one({'BusinessName': 'Penang Flavours'}, {'geocode.latitude','geocode.longitude'}))
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.490142
longitude = 0.08384

penangquery = {'RatingValue': '5',
               'geocode.latitude': {'$lte': (latitude + degree_search), '$gte': (latitude - degree_search)},
               'geocode.longitude': {'$lte': (longitude + degree_search), '$gte': (longitude - degree_search)}}
sort = [('scores.Hygiene', 1)]
limit = 5

# Print the results
for result in establishments.find(penangquery).sort(sort).limit(5):
    pprint(result)
```

```python
# Convert result to Pandas DataFrame
pd.DataFrame(establishments.find(penangquery).sort(sort).limit(5))
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
matchedestablishments = {'$match': {'scores.Hygiene': 0}}

# 2. Groups the matches by Local Authority
groupedmatches = {'$group': {'_id': '$LocalAuthorityName',
                             'count': {'$sum': 1}}}

# 3. Sorts the matches from highest to lowest
sortedmatches = {'$sort': {'count': -1}}

pipeline = [matchedestablishments, groupedmatches, sortedmatches]
matchedresults = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print(f"{len(matchedresults)} documents.")

# Print the first 10 results
pprint(matchedresults[0:10])
```

```python
# Convert the result to a Pandas DataFrame
hygienezeroestablishment = pd.DataFrame(matchedresults)

# Display the number of rows in the DataFrame
print(f"{len(hygienezeroestablishment)} rows are within the hygience score of zero dataframe.")

# Display the first 10 rows of the DataFrame
df.head(10)
```
